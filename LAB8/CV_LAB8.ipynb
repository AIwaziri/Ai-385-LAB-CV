```python
import cv2
import numpy as np

# Load an image in grayscale
gray = cv2.imread(r"C:\Users\User\Downloads\imgs\imgs\chess.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Canny Edge Detection
edges = cv2.Canny(gray, 50, 150)

# Apply Standard Hough Line Transform
lines = cv2.HoughLines(edges, rho=1, theta=np.pi/180, threshold=100)

# Convert edges to color image to draw lines
output = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)

# Draw detected lines
if lines is not None:
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho  # x-coordinate of the line's intersection with the normal
        y0 = b * rho  # y-coordinate of the line's intersection with the normal
        x1 = int(x0 + 1000 * (-b))  # Compute x-coordinates for line extension
        y1 = int(y0 + 1000 * (a))   # Compute y-coordinates for line extension
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(output, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Draw line in red

# Display results
cv2.imshow('Edges', edges)
cv2.imshow('HoughLines - Standard', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load an image in grayscale
gray = cv2.imread(r"C:\Users\User\Downloads\imgs\imgs\building.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(gray, (9, 9), 2)

# Apply Hough Circle Transform
circles = cv2.HoughCircles(blurred, 
                           method=cv2.HOUGH_GRADIENT, 
                           dp=1.2, 
                           minDist=30,
                           param1=50, 
                           param2=30, 
                           minRadius=10, 
                           maxRadius=100)

# Convert to color image for drawing
output = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

# Draw detected circles
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(output, (i[0], i[1]), i[2], (0, 255, 0), 2)  # Draw circle in green
        cv2.circle(output, (i[0], i[1]), 2, (0, 0, 255), 3)  # Draw center in red

# Display results
cv2.imshow('Blurred', blurred)
cv2.imshow('HoughCircles - Detected Circles', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Solve here
import cv2
import numpy as np

# Load the medical image
img = cv2.imread(r"C:\Users\User\Downloads\imgs\imgs\tumor.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 1: Apply Median Blurring to remove noise
# Median blur is often better than Gaussian for medical images as it preserves edges
median_blurred = cv2.medianBlur(gray, 5)

# Step 2: Use Canny Edge Detection to highlight edges
edges = cv2.Canny(median_blurred, 30, 100)

# Step 3: Detect circular boundaries using cv2.HoughCircles()
circles = cv2.HoughCircles(median_blurred, 
                          method=cv2.HOUGH_GRADIENT, 
                          dp=1.2, 
                          minDist=20,
                          param1=50, 
                          param2=30, 
                          minRadius=10, 
                          maxRadius=100)

# Step 4: Draw detected boundaries on the medical image
output = img.copy()
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Draw the outer circle
        cv2.circle(output, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # Draw the center of the circle
        cv2.circle(output, (i[0], i[1]), 2, (0, 0, 255), 3)

# Display all steps for comparison
cv2.imshow('Original Medical Image', img)
cv2.imshow('Median Blurred', median_blurred)
cv2.imshow('Edge Detection', edges)
cv2.imshow('Detected Circular Structures', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
