```python
import cv2 as cv
import numpy as np

cap = cv.VideoCapture(r"C:\Users\4220056\Downloads\vtest.avi")

if not cap.isOpened():
    print("Error: Couldn't open video.")
    exit()

# Get the video frame properties (width, height, frames per second)
frame_width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object to save the output video
fourcc = cv.VideoWriter_fourcc(*'XVID')  # You can use 'MJPG', 'MP4V', etc.
out = cv.VideoWriter(r'C:\Users\4220056\Downloads\optical_flow_output.avi', fourcc, fps, (frame_width, frame_height))

ret, prev_frame = cap.read()

if not ret or prev_frame is None:
    print("Error: Couldn't read the first frame.")
    cap.release()
    exit()

prev_gray = cv.cvtColor(prev_frame, cv.COLOR_BGR2GRAY)

while True:
    ret, next_frame = cap.read()
    if not ret:
        print("Video ended or failed to read frame.")
        break

    next_gray = cv.cvtColor(next_frame, cv.COLOR_BGR2GRAY)
    flow = cv.calcOpticalFlowFarneback(prev_gray, next_gray, None,
                                       pyr_scale=0.5, levels=3, winsize=15,
                                       iterations=3, poly_n=5, poly_sigma=1.2, flags=0)


    mag, ang = cv.cartToPolar(flow[..., 0], flow[..., 1])

    hsv = np.zeros_like(prev_frame)
    hsv[..., 0] = ang * 180 / np.pi / 2  # Hue represents direction
    hsv[..., 1] = 255                    # Full saturation
    hsv[..., 2] = cv.normalize(mag, None, 0, 255, cv.NORM_MINMAX)  # Magnitude

    # Convert HSV to RGB for visualization
    flow_rgb = cv.cvtColor(hsv, cv.COLOR_HSV2RGB)

    # Write the optical flow frame to the output video (only the flow, no original frame)
    out.write(flow_rgb)

    # Update previous frame for the next iteration
    prev_gray = next_gray

# Release the video capture and output objects
cap.release()
out.release()

print("Optical flow video saved as 'output.avi'")
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

cap = cv.VideoCapture(r"C:\Users\4220056\Downloads\vtest.avi")

if not cap.isOpened():
    print("Error: Couldn't open video.")
    exit()

frame_width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv.CAP_PROP_FPS)

fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter(r'C:\Users\4220056\Downloads\optical_flow_output.avi', fourcc, fps, (frame_width, frame_height))

ret, prev_frame = cap.read()

if not ret or prev_frame is None:
    print("Error: Couldn't read the first frame.")
    cap.release()
    exit()

prev_gray = cv.cvtColor(prev_frame, cv.COLOR_BGR2GRAY)

prev_pts = cv.goodFeaturesToTrack(prev_gray, maxCorners=100, qualityLevel=0.3, minDistance=7)

# Assigning unique colors to each point (use a fixed color palette)
colors = [tuple(np.random.randint(0, 255, 3).tolist()) for _ in range(100)]

mask = np.zeros_like(prev_frame)

plt.ion()
fig, ax = plt.subplots()

while True:
    ret, next_frame = cap.read()
    if not ret:
        print("Video ended or failed to read frame.")
        break

    next_gray = cv.cvtColor(next_frame, cv.COLOR_BGR2GRAY)

    next_pts, status, err = cv.calcOpticalFlowPyrLK(prev_gray, next_gray, prev_pts, None, winSize=(21, 21), maxLevel=3,
                                                    criteria=(cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 30, 0.01))

    good_new = next_pts[status == 1]
    good_old = prev_pts[status == 1]

    for i, (new, old) in enumerate(zip(good_new, good_old)):
        a, b = new.ravel()
        c, d = old.ravel()

        color = colors[i]  # Use a fixed color from the palette

        mask = cv.line(mask, (int(a), int(b)), (int(c), int(d)), color, 2)
        next_frame = cv.circle(next_frame, (int(a), int(b)), 5, color, -1)

    overlay = cv.addWeighted(next_frame, 0.7, mask, 0.3, 0)

    overlay_rgb = cv.cvtColor(overlay, cv.COLOR_BGR2RGB)

    ax.clear()

    ax.imshow(overlay_rgb)
    ax.axis('off')

    plt.pause(0.01)

    out.write(overlay)

    prev_gray = next_gray
    prev_pts = good_new.reshape(-1, 1, 2)

plt.ioff()
plt.show()

cap.release()
out.release()

print("Optical flow video saved as 'optical_flow_output.avi'")
```
