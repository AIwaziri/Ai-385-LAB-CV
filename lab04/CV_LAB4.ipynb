# 1-	Sobel Operator (Edge Detection):

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

image1 = cv2.imread("/content/Img 3.jpeg")
image2 = cv2.imread("/content/Img 2.jpeg")
image3 = cv2.imread("/content/Img 1.jpeg")
```

```python
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
sobelx = cv2.Sobel(src = gray_image1 , ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 5)
sobely = cv2.Sobel(src = gray_image1 , ddepth = cv2.CV_64F, dx = 0, dy = 1, ksize = 5)
sobelxy = cv2.Sobel(src = gray_image1 , ddepth = cv2.CV_64F, dx = 1, dy = 1, ksize = 5)

cv2_imshow(sobelx)
cv2.waitKey(0)
cv2_imshow(sobely)
cv2.waitKey(0)
cv2_imshow(sobelxy)
cv2.waitKey(0)
```

# 2-	Canny Edge Detector:

```python
threshold1 = 50
threshold2 = 150

MeEdges = cv2.Canny(image1, threshold1, threshold2)

cv2_imshow(MeEdges)
```

# 3-	Harris Corner Detector:

```python
img = image1

gray_image1 = np.float32(gray_image1)

harris_corners = cv2.cornerHarris(gray_image1, blockSize=2, ksize=3, k=0.04)
harris_corners = cv2.dilate(harris_corners, None)

img[harris_corners > 0.01 * harris_corners.max()] = [0, 0, 255]

cv2_imshow(img)
```
